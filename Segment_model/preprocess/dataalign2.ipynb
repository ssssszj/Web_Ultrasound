```python
import cv2
import xml.etree.ElementTree as ET
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve, auc
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import random
```

```python
new_machine = {}
machine_new = {}
tuomin = pd.read_csv('D:/Lab/ultrasound PTB/data/second_data/annotations/second anno/20220311-cervix-tuomin/脱敏对应.csv')
for t in tuomin.itertuples():
    machine = t[1].split('/')[-1]
    new = t[2].split('/')[-1]
    new_machine[new] = machine
    machine_new[machine] = new

def is_chinese(strings):
    for _char in strings:
        if '\u4e00' <= _char <= '\u9fa5':
            return True

ori_dir = 'D:/Lab/ultrasound PTB/data/second_data/rawdata/pictures/'
machine_name = {}

for subdir in os.listdir(ori_dir):
    for filename in os.listdir(os.path.join(ori_dir,subdir)):
        name = ''
        for char in filename:
            if is_chinese(char):
                name += char
            else:
                break
        if os.path.isdir(os.path.join(ori_dir,subdir,filename)):
            for jpgname in os.listdir(os.path.join(ori_dir,subdir,filename)):
                if jpgname[-4:] != '.jpg':
                    continue
                else:
                    machine_name[jpgname] = name
        else:
            if filename[-4:] != '.jpg':
                continue
            else:
                machine_name[filename] = name
    
```

```python
news = []
machines = []
names = []
labels = []
cervixs = []
polygons = []
masks = []
img_paths = []
base_dir = 'D:/Lab/ultrasound PTB/data/second_data/annotations/second anno/20220311-cervix-tuomin/cervix-tuomin/'
root = ET.parse('D:/Lab/ultrasound PTB/data/second_data/annotations/second anno/annotations.xml').getroot()
df = pd.read_excel('D:/Lab/ultrasound PTB/data/second_data/rawdata/Excel/2022-02-部分数据.xlsx',index_col=1)

for image in root.findall('image'):
    flag=True
    path = image.attrib['name']
    
    new = path.split('/')[-1]
    img_path = os.path.join(base_dir,new)
    machine = new_machine[new]
    name = machine_name[machine]
    try:
        label = float(df.loc[name]['PTB'])
        cervix = df.loc[name]['孕中期宫颈长度1']
        print(name,cervix)
        if np.isnan(cervix):
            flag = False
    except:
        flag = False
    for anno in image:
        if anno.tag == 'polygon':
            polygon = np.array(list(map(
                lambda x: np.array(x.strip().split(','), dtype=np.float32),
                anno.attrib['points'].split(';')
            )))
            mask =  np.around(polygon).astype(np.int32)

    
    if flag:
        news.append(new)
        machines.append(machine)
        names.append(name)
        labels.append(label)
        cervixs.append(cervix)
        polygons.append(polygon)
        masks.append(mask)
        img_paths.append(img_path)
    
print(len(news)) 
print(len(labels))
```

```python
root = ET.parse('D:/Lab/ultrasound PTB/data/second_anno/annotations.xml').getroot()
base_dir = 'D:/Lab/ultrasound PTB/data/Dataset/Annotations/'

for image in root.findall('image'):
    path = image.attrib['name']
    filename = path.split('/')[-1][:-4]+'.png'
    img_path = os.path.join(base_dir,filename)

    width = int(image.attrib['width'])
    height = int(image.attrib['height'])

    for anno in image:
        if anno.tag == 'polygon':
            polygon = np.array(list(map(
                lambda x: np.array(x.strip().split(','), dtype=np.float32),
                anno.attrib['points'].split(';')
            )))
            mask = np.around(polygon).astype(np.int32)
    canvas = np.zeros((height,width))
    cv2.fillPoly(canvas, [mask], 1)
    cv2.imwrite(img_path,canvas)
```
